**Noise Removal Code**


```
noise_list = ["is", "a", "this", "..."] 
def _remove_noise(input_text):
    words = input_text.split() 
    noise_free_words = [word for word in words if word not in noise_list] 
    noise_free_text = " ".join(noise_free_words) 
    return noise_free_words

_remove_noise("this is a sample text")
```

**REMOVAL OF HASHTAGS**


```
import re 

def _remove_regex(input_text, regex_pattern):
    # MATCHING INPUT TEXT WITH PATTERN
    urls = re.finditer(regex_pattern, input_text) 
    for i in urls: 
        # Subtituting the sentence
        input_text = re.sub(i.group().strip(), '', input_text)
    return input_text
# Now We declare pattern
regex_pattern = "#[\w]*"  

_remove_regex("remove this #hashtag from analytics vidhya", regex_pattern)
```

**LEMMATIZING WORD**

**Object standardizing**

some of the words are written in short form cant be understood by models

```
#Dictionery of short form and their meaning
lookup_dict = {'rt':'Retweet', 'dm':'direct message', "awsm" : "awesome", "luv" :"love"}
def _lookup_words(input_text):
    #splitting each word from given sentance
    words = input_text.split() 
    new_words = [] 
    for word in words:
        #Finding the dictonery words in input text
        if word.lower() in lookup_dict:
            word = lookup_dict[word.lower()]
        new_words.append(word)
    new_text = " ".join(new_words)
    return new_text

_lookup_words("RT this is a retweeted tweet by Shivam Bansal")
```

**PART OF SPEECH**



```
from nltk import word_tokenize, pos_tag
text = "I am learning Natural Language Processing on Analytics Vidhya"
#CREATING TOKENS FROM TEXT
tokens = word_tokenize(text)
#Printing in as partof speech for each word
print(pos_tag(tokens))
```

```
doc1 = "Sugar is bad to consume. My sister likes to have sugar, but not my father." 
doc2 = "My father spends a lot of time driving my sister around to dance practice."
doc3 = "Doctors suggest that driving may cause increased stress and blood pressure."
doc_complete = [doc1, doc2, doc3]
doc_clean = [doc.split() for doc in doc_complete]

import gensim
import corpora

# Creating the term dictionary of our corpus, where every unique term is assigned an index.  
dictionary = corpora.Dictionary(doc_clean)

# Converting list of documents (corpus) into Document Term Matrix using dictionary prepared above. 
doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]

# Creating the object for LDA model using gensim library
Lda = gensim.models.ldamodel.LdaModel

# Running and Training LDA model on the document term matrix
ldamodel = Lda(doc_term_matrix, num_topics=3, id2word = dictionary, passes=50)

# Results 
print(ldamodel.print_topics())
```

```
def generate_ngrams(text, n):
    words = text.split()
    output = []  
    for i in range(len(words)-n+1):
        output.append(words[i:i+n])
    return output
generate_ngrams('this is a sample text which is better to understand', 3)
```
